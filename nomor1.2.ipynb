```python
import simpy
import matplotlib.pyplot as plt
```

```python
def traffic_light(env, light_states):
    while True:
        # Lampu merah
        print(f"Lampu lalu lintas berubah menjadi Merah pada waktu {env.now:.2f} detik")
        light_states.append(('RED', env.now))
        yield env.timeout(30)  

        # Lampu hijau
        print(f"Lampu lalu lintas berubah menjadi Hijau pada waktu {env.now:.2f} detik")
        light_states.append(('GREEN', env.now))
        yield env.timeout(45) 

        # Lampu kuning
        print(f"Lampu lalu lintas berubah menjadi Kuning pada waktu {env.now:.2f} detik")
        light_states.append(('YELLOW', env.now))
        yield env.timeout(5) 
```

### **Menjalankan Proses Simulasi**

```python
env = simpy.Environment()
lampu_lalu_lintas = []
env.process(traffic_light(env, lampu_lalu_lintas))
env.run(until=180)
```

### **Perhitungan Waktu**
Untuk siklus lampu:

1. Lampu merah: 30 detik
2. Lampu hijau: 45 detik
3. Lampu kuning: 5 detik

Total durasi satu siklus adalah 30+45+5 = 80 detik. 

Jadi, dalam 180 detik, kita bisa memasukkan:

180 : 80 â‰ˆ 2.25
Ini berarti hanya ada waktu untuk sekitar 2 siklus penuh (160 detik) sebelum simulasi dihentikan.

### **Visualisasi data dengan matplotlib**

```python
plt.figure(figsize=(10, 5))
for i in range(len(lampu_lalu_lintas) - 1):
    plt.plot([lampu_lalu_lintas[i][1], lampu_lalu_lintas[i + 1][1]], [i%3, i%3], color=lampu_lalu_lintas[i][0], linewidth=10)

plt.yticks([0, 1, 2], ['Merah', 'Hijau', 'Kuning'])
plt.xlabel('Waktu (detik)')
plt.title('Simulasi Lampu Lalu Lintas')
plt.grid(True)
plt.show()
```
