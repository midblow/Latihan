### **Hipotesis**

> H1 (Hipotesis Alternatif): Terdapat hubungan antara variabel-variabel seperti durasi tidur, jenis pekerjaan, jenis kelamin, dan umur dengan kemungkinan seseorang mengalami gangguan tidur seperti sleep apnea atau insomnia.

> H0 (Hipotesis Nol): Tidak terdapat hubungan antara variabel-variabel seperti durasi tidur, jenis pekerjaan, jenis kelamin, dan umur dengan kemungkinan seseorang mengalami gangguan tidur seperti sleep apnea atau insomnia.

### **Alasan Penggunaan Metode**

> Metode Naive Bayes cocok digunakan pada kasus ini karena beberapa alasan:
> 1. Kemudahan Implementasi: Naive Bayes adalah metode yang relatif sederhana dan mudah diimplementasikan, terutama ketika digunakan dalam klasifikasi. Hal ini membuatnya cocok untuk proyek seperti ini yang membutuhkan solusi yang cepat dan efisien.
> 2. Efisien dalam Penanganan Data Besar: Meskipun dataset tidak dijelaskan, Naive Bayes biasanya efisien dalam menangani data besar, yang mungkin diperlukan untuk analisis gangguan tidur karena banyaknya variabel yang mungkin relevan.
> 3. Kemampuan Menangani Variabel Diskrit dan Kontinu: Naive Bayes dapat menangani variabel diskrit dan kontinu dengan baik, seperti durasi tidur (kontinu) dan jenis pekerjaan (diskrit). Ini penting karena dataset mungkin mengandung beragam jenis data.
> 4. Kemampuan Menangani Data yang Tidak Seimbang: Jika data mengandung kelas yang tidak seimbang (misalnya, jumlah individu yang menderita gangguan tidur lebih sedikit daripada yang tidak), Naive Bayes tetap dapat memberikan klasifikasi yang akurat.
> 5. Kemampuan Mengatasi Ketergantungan Antara Variabel: Meskipun metode ini memiliki asumsi naif tentang independensi antar variabel, seringkali masih memberikan hasil yang baik dalam praktik, terutama jika asumsi tersebut hampir terpenuhi atau jika ketergantungan antar variabel tidak terlalu kuat.

### **Mendapatkan Nilai Probabilitas untuk Setiap Jenis Sleep Disorder**

> `def probabilitySleepDisorder():` merupakan deklarasi fungsi yang bertujuan untuk menghitung dan mengembalikan jumlah serta probabilitas dari setiap jenis gangguan tidur dalam dataset. Fungsi ini mengembalikan tiga hasil utama:
> - `result_list` berupa daftar tuple yang memuat jenis gangguan tidur, jumlah kemunculan, dan probabilitas
> - `sleep_disorder_counts` berupa series yang memuat jumlah kemunculan setiap jenis gangguan tidur 
> - `sleep_disorder_prob` berupa series yang memuat probabilitas setiap jenis gangguan tidur.

> Fungsi ini menggunakan metode `value_counts()` dari pustaka pandas untuk menghitung jumlah kemunculan setiap jenis gangguan tidur dalam kolom Sleep Disorder. Jumlah total entri dalam dataset dihitung menggunakan fungsi `len()`. Hasilnya adalah jumlah total data dalam dataset yang kemudian dibagi dengan jumlah total data untuk mendapatkan probabilitas dari setiap jenis gangguan tidur.

```python
def probabilitySleepDisorder():
    sleep_disorder_counts = dataset['Sleep Disorder'].value_counts()
    total_count = len(dataset)
    sleep_disorder_prob = sleep_disorder_counts/total_count
    result_list = [
    (sleep_disorder, count, sleep_disorder_prob[sleep_disorder])
    for sleep_disorder, count in sleep_disorder_counts.items() 
    ]

    return result_list, sleep_disorder_counts, sleep_disorder_prob
```

> Pada baris ini, fungsi `probabilitySleepDisorder()` dipanggil dan hasilnya disimpan dalam tiga variabel, yaitu: variabel `result` berupa daftar yang berisi tuple dari jenis gangguan tidur, jumlah kasus, dan probabilitas, lalu variabel `sleep_disorder_counts` berupaeries yang berisi jumlah kasus setiap jenis gangguan tidur, dan variabel`sleep_disorder_prob` berupa series yang berisi probabilitas setiap jenis gangguan tidur. Kemudian menggunakan for loop untuk mengiterasi setiap elemen dalam result. Setiap elemen adalah sebuah tuple yang memuat informasi tentang jenis gangguan tidur `(i)`, jumlah kasus `(j)`, dan probabilitas `(k)`. Pada setiap iterasi, kode ini mencetak informasi tersebut ke layar.

```python
result, sleep_disorder_counts, sleep_disorder_prob = probabilitySleepDisorder()
for i, j, k in result:
    print("Sleep Disorder:", i)
    print("Count:", j)
    print("Probability:", k)
    print()
```

### **Membuat Function Visualiasasi Data**

> `def printTable(table, tipe):` merupakan deklarasi fungsi yang digunakan untuk menampilkan data dalam bentuk tabel. Fungsi ini menerima dua parameter berupa `table` yaitu DataFrame yang akan ditampilkan dan `tipe` yaitu sebuah string yang menjelaskan tipe data yang sedang ditampilkan (misalnya, 'Gender', 'Occupation', dll.). Fungsi ini membuat sebuah figure baru dengan ukuran 10x0.2 inci dan menambahkan subplot tunggal di figure tersebut. Metode `ax.table` digunakan untuk menambahkan tabel ke subplot dengan data dari DataFrame table digunakan sebagai isi tabel (`cellText`), dengan label baris (`rowLabels`) dan label kolom (`colLabels`). Atribut cellLoc diatur ke 'center' untuk menempatkan teks di tengah sel.

> `def plot_sleep_disorder_distribution(column):` merupakan deklarasi fungsi yang digunakan untuk membuat plot distribusi gangguan tidur berdasarkan kolom tertentu. Fungsi ini menerima satu parameter yaitu `column` yang berupa nama kolom dalam dataset yang akan digunakan untuk membuat plot distribusi (misalnya, 'Gender', 'Occupation', 'Age', atau 'Sleep Duration'). Fungsi ini membuat figure baru dengan ukuran 10x10 inci untuk plot, lalu memeriksa apakah kolom yang diberikan termasuk dalam daftar kolom yang valid untuk visualisasi. Kode `sns.set_style` digunakan untuk mengatur gaya plot. Kemudian membuat histogram tumpuk `(multiple='stack')` dengan `sns.histplot` berdasarkan data dalam dataset. Kolom yang diatur oleh parameter column digunakan sebagai sumbu x, dan warna plot ditentukan oleh nilai dalam kolom Sleep Disorder nya. Khusus untuk kolom `Occupation`, maka teks pada sumbu x akan diputar sebesar 45 derajat untuk meningkatkan keterbacaan. Jika kolom yang diberikan tidak valid, maka fungsi akan mencetak pesan kesalahan dan keluar dari fungsi tanpa menampilkan plot.

```python
def printTable(table, tipe):
    fig = plt.figure(figsize = (10, .2))
    ax = fig.add_subplot(111)
    ax.table(cellText = table.values, rowLabels = table.index, colLabels = table.columns, cellLoc = 'center')
    ax.set_title(f"Jumlah Data Sleep Disorder Berdasarkan {tipe}")
    ax.axis('off')

def plot_sleep_disorder_distribution(column):
    plt.figure(figsize=(10, 10))
    if column in ['Gender', 'Occupation', 'Age', 'Sleep Duration']:
        sns.set_style("whitegrid")
        sns.histplot(data=dataset, x=column, hue='Sleep Disorder', multiple='stack', kde=False, palette={'None': 'lavender', 'Sleep Apnea': 'cornflowerblue', 'Insomnia': 'royalblue'}, edgecolor=None, linewidth=1, alpha=0.7)
        plt.title(f'Distribusi Data Berdasarkan {column}')
        plt.xlabel(column)
        plt.ylabel('Count')
        
        if column == 'Occupation':
            plt.xticks(rotation=45) 
    else:
        print(f"Column '{column}' is not valid for this visualization.")
        return
    
    plt.show()
```

Sedangkan jika `sleepDur i` tidak ditemukan dalam index, probabilitas diatur ke 0. Setelah dilakukan perhitungan kemudian probabilitas yang dihitung ditambahkan ke dalam list yang sesuai dalam dictionary `sleepDur_prob`.

> Kemudian hasil perhitungan ditampilkan dengan menggunakan `for loop` bersarang untuk melalui setiap kategori gangguan tidur. Looping melalui setiap `sleepDur` dan probabilitas yang sesuai dan mencetak probabilitas bersyarat untuk `sleepDur i` dan kategori gangguan tidur category dengan tiga digit desimal.
```python
sd_category = ['None', 'Sleep Apnea', 'Insomnia']
sleepDur = by_sleep_duration.index
sleepDur_prob = {category: [] for category in sd_category}

for category in sd_category:    
    for i in sleepDur:
        if i in sleep_disorder_by_sleep_duration.index:
            prob = sleep_disorder_by_sleep_duration.loc[i, category] / sleep_disorder_counts[category]
        else:
            prob = 0
        sleepDur_prob[category].append(prob)

for category in sd_category:
    print(f"P(Sleep Duration | {category}):")
    for i, j in zip(sleepDur, sleepDur_prob[category]):
        print(f"P(Sleep Duration = {i} | {category}) = {j:.3f}")
    print()
```

### **Setelah didapatkan Hasil Ujicoba Sampel Selanjutnya Melakukan Perhitungan Bayes Terhadap Keseluruhan Data**

> `def hitung_probabilitas(dataset):` merupakan deklrasi fungsi yang akan digunakan untuk menghitung probabilitas dengan memasukkan parameter berupa `dataset` yang akan dihitung. Fungsi ini pertama-tama akan menghitung jumlah setiap jenis gangguan tidur dalam dataset dan disimpan ke dalam variabel `sd_total` dan juga menghitung probabilitas dasar dari setiap gangguan tidur dengan smoothing Laplace untuk menghindari probabilitas nol dan disimpan ke dalam variabel `sd_prob`.  Kemudian dilakukan perhitungan probabilitas bersyarat untuk setiap kategori yang dihitung yaitu Gender, Occupation, Age, dan Sleep Duration, dengan menambahkan 1 pada jumlah setiap kategori dan membaginya dengan total jumlah ditambah jumlah unik dari kategori tersebut. Fungsi ini mengembalikan probabilitas dasar gangguan tidur dan probabilitas bersyarat untuk setiap fitur (Gender, Occupation, Age, Sleep Duration) yang telah diperoleh.

> `prob_all, gender_all, occ_all, age_all, sleep_duration_all = hitung_probabilitas(dataset)` merupakan pemanggilan fungsi `hitung_probabilitas()` untuk menghitung nilai probabilitas yang hasil pengembaliannya akan disimpan ke dalam masing-masing variabel.

> `def conditional_prob(case, kategori, probabilitas_bersyarat, sd_jumlah):` merupakan deklrasi fungsi yang akan digunakan untuk menghitung probabilitas bersyarat, yang apabila case (misalnya Gender tertentu) ada dalam index `probabilitas_bersyarat`, maka probabilitasnya diambil dari `probabilitas_bersyarat` tersebut. Sedangkan jika tidak, maka probabilitas dihitung dengan smoothing Laplace untuk menghindari pembagian dengan nol.

> `def prediksi_sd(case):` merupakan deklrasi fungsi yang akan digunakan untuk memprediksi jenis gangguan tidur (Sleep Disorder) berdasarkan data dari seorang individu dengan menggunakan model Naive Bayes dengan probabilitas bersyarat yang telah dihitung sebelumnya. Fungsi ini menerima satu parameter, yaitu `case` yang berisi data individu yang akan diprediksi. Pertama-tama diinisialisasikan probabilitas dasar dan probabilitas bersyarat untuk setiap fitur yang telah dihitung menggunakan smoothing Laplace. Lalu dideklarasikan list `sd_kategori` yang menyimpan 'None', 'Sleep Apnea', 'Insomnia', serta `conditional_probs` untuk menyimpan probabilitas bersyarat setiap kategori gangguan tidur. Kemudian dilakukan perulangan dengan `for loop` melalui setiap kategori gangguan tidur dan dengan menggunakan fungsi `conditional_prob()` untuk menghitung probabilitas bersyarat dari Gender, Occupation, Age, dan Sleep Duration untuk setiap kategori gangguan tidur dan menyimpan hasil probabilitas bersyaratnya dalam `conditional_probs`. Lalu dihitung probabilitas total dengan Naive Bayes menggunakan loop untuk mengambil  probabilitas bersyarat yang telah dihitung dan menghitung probabilitas total menggunakan rumus Naive Bayes: P(Sleep Disorder | Features) = P(Sleep Disorder) * P(Gender | Sleep Disorder) * P(Occupation | Sleep Disorder) * P(Age | Sleep Disorder) * P(Sleep Duration | Sleep Disorder). Setelah diperoleh hasil dari perhitungan Naive Bayes kemudian mengitung total probabilitas dan memastikan bahwa totalnya 1. Jika total probabilitas lebih dari 0, normalisasi setiap probabilitas dengan membagi totalnya. List `simpan_hasil` digunakan untuk menyimpan hasil probabilitas untuk setiap kategori gangguan tidur dan membuat string yang berisi hasil probabilitas dengan format yang mudah dibaca. Terakhir adalah menentukan hasil prediksi jenis gangguan tidur dengan probabilitas tertinggi yang diperoleh dengan menggunakan fungsi max dan menyimpan hasil prediksi dalam `prediksi_output` sebagai return value dari fungsi ini.


```python
# Fungsi untuk menghitung probabilitas dengan smoothing Laplace
def hitung_probabilitas(dataset):
    sd_total = dataset['Sleep Disorder'].value_counts()
    sd_prob = (sd_total + 1) / (len(dataset) + len(sd_total))

    sleep_disorder_by_gender = (dataset.groupby(['Gender', 'Sleep Disorder']).size().unstack(fill_value=0) + 1) / \
                            (sd_total + len(dataset['Gender'].unique())).reindex(sd_total.index, fill_value=0)
    
    sleep_disorder_by_occupation = (dataset.groupby(['Occupation', 'Sleep Disorder']).size().unstack(fill_value=0) + 1) / \
                            (sd_total + len(dataset['Occupation'].unique())).reindex(sd_total.index, fill_value=0)
    
    sleep_disorder_by_age = (dataset.groupby(['Age', 'Sleep Disorder']).size().unstack(fill_value=0) + 1) / \
                            (sd_total + len(dataset['Age'].unique())).reindex(sd_total.index, fill_value=0)
    
    sleep_disorder_by_sleep_duration = (dataset.groupby(['Sleep Duration', 'Sleep Disorder']).size().unstack(fill_value=0) + 1) / \
                            (sd_total + len(dataset['Sleep Duration'].unique())).reindex(sd_total.index, fill_value=0)
    
    return sd_prob, sleep_disorder_by_gender, sleep_disorder_by_occupation, sleep_disorder_by_age, sleep_disorder_by_sleep_duration

# Menghitung probabilitas untuk seluruh dataset
prob_all, gender_all, occ_all, age_all, sleep_duration_all = hitung_probabilitas(dataset)

# Fungsi untuk menghitung probabilitas bersyarat
def conditional_prob(case, kategori, probabilitas_bersyarat, sd_jumlah):
    if case in probabilitas_bersyarat.index:
        return probabilitas_bersyarat.loc[case, kategori]
    else:
        return 1 / (sd_jumlah[kategori] + len(probabilitas_bersyarat))

# Fungsi prediksi sleep disorder
def prediksi_sd(case):
    sd_prob = prob_all
    sd_by_gender = gender_all
    sd_by_occupation = occ_all
    sd_by_age = age_all
    sd_by_sleep_duration = sleep_duration_all

    sd_kategori = ['None', 'Sleep Apnea', 'Insomnia']
    conditional_probs = {}

    for kategori in sd_kategori:
        P_gender_sd = conditional_prob(case['Gender'], kategori, sd_by_gender, sd_prob)
        P_occupation_sd = conditional_prob(case['Occupation'], kategori, sd_by_occupation, sd_prob)
        P_age_sd = conditional_prob(case['Age'], kategori, sd_by_age, sd_prob)
        P_sleep_duration_sd = conditional_prob(case['Sleep Duration'], kategori, sd_by_sleep_duration, sd_prob)
        
        conditional_probs[kategori] = (P_gender_sd, P_occupation_sd, P_age_sd, P_sleep_duration_sd)

    # Rumus Naive Bayes : P(Sleep Disorder | Features) = P(Sleep Disorder) * P(Gender | Sleep Disorder) * P(Occupation | Sleep Disorder) * P(Age | Sleep Disorder) * P(Sleep Duration | Sleep Disorder)
    naiveBayes = {}

    for kategori in sd_kategori:
        P_gender_sd, P_occupation_sd, P_age_sd, P_sleep_duration_sd = conditional_probs[kategori]
        nb = sd_prob[kategori] * P_gender_sd * P_occupation_sd * P_age_sd * P_sleep_duration_sd
        naiveBayes[kategori] = nb

    # Normalisasi probabilitas
    total_prob = sum(naiveBayes.values())
    if total_prob > 0:
        for kategori in naiveBayes:
            naiveBayes[kategori] /= total_prob

    # Print hasil probabilitas
    simpan_hasil = []
    for kategori, prob in naiveBayes.items():
        hasil_prob = f"P({kategori} | Gender = {case['Gender']}, Occupation = {case['Occupation']}, Age = {case['Age']}, Sleep Duration = {case['Sleep Duration']}) = {prob:.5f}"
        simpan_hasil.append(hasil_prob)

    # Menentukan hasil sleep disorder dengan probabilitas tertinggi dari hasil_prob
    sd_prediksi = max(naiveBayes, key=naiveBayes.get)
    prediksi_output = f"Hasil Prediksi sleep disorder pada kasus ini adalah: {sd_prediksi}"
    
    return simpan_hasil, prediksi_output
```

> `def visualisasi_prediksi(results, case, case_number):` merupakan deklarasi fungsi yang digunakan untuk memvisualisasikan hasil prediksi jenis gangguan tidur (Sleep Disorder) dalam bentuk diagram batang. Fungsi ini menerima tiga parameter, yaitu:    `results` (list yang berisi string hasil probabilitas dari fungsi prediksi_sd), `case` (dictionary yang berisi data individu yang akan diprediksi), dan `case_number` (nomor kasus yang akan divisualisasikan). Pertama-tama dilakukan inisialisasi kategori gangguan tidur `sleep_disorder = ['None', 'Sleep Apnea', 'Insomnia']` dan juga warna plot menggunakan pustaka seaborn engan menggunakan palet `husl`. Kemudian dibuat figure dengan `fig, ax = plt.subplots(figsize=(12, 6))` untuk plot dengan ukuran 12x6 inci. Kemudian mengekstrak probabilitas dari hasil prediksi yang diperoleh yang akan disimpan ke dalam `probs` yang diinisialisasi dengan nilai 0. Dilakukan looping melalui setiap hasil prediksi dalam results untuk mengekstrak kategori dan probabilitas. Variabel `prob_start_idx` menentukan posisi awal nilai probabilitas dalam string hasil prediksi, sedangkan variabel `kategori_end_idx` menentukan posisi akhir nama kategori dalam string hasil prediksi. Kemudian mengambil nama kategori dari string hasil prediksi dan mengambil nilai probabilitas dari string hasil prediksi dan mengkonversinya menjadi float, lalu menyimpan probabilitas dalam probs untuk kategori yang sesuai.

> Setelah itu memvisualisasikan hasil prediksi ke dalam suatu diagram batang dengan menggunakan `seaborn` untuk membuat diagram batang dari probabilitas dengan x sebagai kategori dan y sebagai nilai probabilitas, dan menentukan batas atas sumbu y menjadi 1 untuk probabilitas.

```python
# Fungsi visualisasi hasil prediksi
def visualisasi_prediksi(results, case, case_number):
    sleep_disorder = ['None', 'Sleep Apnea', 'Insomnia']
    x = np.arange(len(sleep_disorder))
    colors = sns.color_palette("husl", len(sleep_disorder))

    fig, ax = plt.subplots(figsize=(12, 6))

    # Ekstrak probabilitas dari hasil prediksi
    probs = {kategori: 0 for kategori in sleep_disorder}
    for result in results:
        prob_start_idx = result.rfind('= ') + 2
        kategori_end_idx = result.find(' | ')
        kategori = result[2:kategori_end_idx]
        prob = float(result[prob_start_idx:])
        probs[kategori] = prob

    # Plot probabilitas
    sns.barplot(x=list(probs.keys()), y=list(probs.values()), ax=ax, palette=colors)
    ax.set_ylim(0, 1)
    ax.set_title(f'Case {case_number}: {case}', fontsize=14)
    ax.set_ylabel('Probabilitas', fontsize=12)
    ax.set_xlabel('Sleep Disorder Kategori', fontsize=12)
    for p in ax.patches:
        ax.annotate(format(p.get_height(), '.2f'),
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center',
                    xytext=(0, 9),
                    textcoords='offset points')

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
```

> `def visualisasi_distribusi_normalisasi(column):` merupakan deklarasi fungsi yang digunakan untuk melakukan langkah-langkah memeriksa apakah data dalam kolom tertentu mengikuti distribusi normal. Ini termasuk normalisasi data, perhitungan Z-score, PDF, dan korelasi, serta visualisasi hasil dalam bentuk scatter plot dan histogram. Jika kolom sesuai maka data dari kolom yang diberikan diambil dari dataset dan dimasukkan ke dalam DataFrame baru bernama `normalisasi`. Kemudian dilakukan perhitungan `rank` (diberi peringkat dari yang terkecil hingga terbesar), `indeks` (berdasarkan rumus yang digunakan untuk normalisasi), `Formula Z-Score` (Menggunakan distribusi normal standar (mean = 0, std = 1), fungsi ini menghitung nilai Z-score berdasarkan indeks yang dihitung sebelumnya), dan `PDF` atau Probability Density Function untuk setiap nilai Z-score yang diperoleh. Dihitung juga korelasi antara data asli (Contoh Data) dan nilai Z-score dihitung untuk melihat seberapa dekat data mengikuti distribusi normal dan menampilkab hasil normalisasi dan korelasi yang diperoleh.

```python
def visualisasi_distribusi_normalisasi(column):
    if column in ['Age', 'Sleep Duration']:
        normalisasi = pd.DataFrame({
            'Contoh Data': dataset[column]
        })

        normalisasi['Rank'] = normalisasi['Contoh Data'].rank(method='min', ascending=True).astype(int)
        normalisasi['Indeks'] = (normalisasi['Rank'] - 3/10) / (len(normalisasi) + 1/4)
        normalisasi['Formula Z-Score'] = norm.ppf(normalisasi['Indeks'], 0, 1)
        normalisasi['PDF'] = norm.pdf(normalisasi['Formula Z-Score'], loc=0, scale=1)
        corr = normalisasi['Contoh Data'].corr(normalisasi['Formula Z-Score'])

        print(normalisasi)
        print("Correlation = ", corr)

        plt.scatter(normalisasi['Formula Z-Score'], normalisasi['PDF'], color='purple')
        plt.title(f'Scatter Plot PDF untuk Kolom {column}')
        plt.xlabel('Formula Z-Score')
        plt.ylabel('PDF')
        plt.grid(True)
        plt.show()

        plt.hist(normalisasi['Formula Z-Score'], color='purple', edgecolor='pink', linewidth=1, alpha=0.7)
        plt.title(f'Histogram untuk Uji Normalitas pada Kolom {column}')
        plt.xlabel('Formula Z-Score')
        plt.ylabel('Count')
        plt.grid(True)
        plt.show()
    else:
        print(f"Column '{column}' is not valid for this visualization.")
        return

columns_to_test = ['Age', 'Sleep Duration']
for column in columns_to_test:
    visualisasi_distribusi_normalisasi(column)
```
